```python
import pandas as pd

import re
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_pickle("ce_data.pkl")
```

```python
df =df.reset_index(drop=True)
```

```python
def clean(sen):
    sen = re.sub('[^a-zA-Z0-9 \n\.]', ' ', sen)
    sen = sen.replace('.', ' ')
    sen = sen.lower()
    sen = sen.split()
    return sen
```

```python
token = []
for i in df['sentences']:
    cleaned = clean(i)
    token.append(cleaned)
```

```python
df['token'] = token
```

```python
label = []
for index,row in df.iterrows():
    tokens = row['token']
    cause = row['cause'].lower().split()
    effect = row['effect'].lower().split()
    labels_tok = []
    for i in tokens:
        if i in cause:
            labels_tok.append('C')
        elif i in effect:
            labels_tok.append('E')
        else:
            labels_tok.append('O')
    label.append(labels_tok)
    
    
    
df['label'] = label    
```

```python
df
```

```python
train , test = train_test_split(df, test_size=0.2, random_state=25)
```

```python
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
```
