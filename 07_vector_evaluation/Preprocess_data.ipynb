```python
import pandas as pd
import re
```

```python
data = pd.read_csv("TRAIN_FILE.TXT" , delimiter= '\t', header=None)
```

```python
data = data.rename({0: 'sen_id', 1: 'sentences', 2 : 'Relation_type' , 3: 'Comments'}, axis=1)
```

```python
data
```

```python
ce_data = data[data['Relation_type'].str.contains('Cause-Effect')]
```

```python
ce_data
```

```python
cause = []
effect = []
relation = []

for index,row in ce_data.iterrows():
    sen = row['sentences']
    rel = row['Relation_type']
    
    e1 = re.search(r'<e1\>(.*?)\</e1>', sen).group(1)
    e2 = re.search(r'<e2\>(.*?)\</e2>', sen).group(1)
    
    if rel== "Cause-Effect(e2,e1)":
        cause.append(e2)
        effect.append(e1)
        relation.append((e2,e1))

        
    else:
        cause.append(e1)
        effect.append(e2)
        relation.append((e1,e2))
    sen = sen.replace("<e2>", " ")
    sen = sen.replace("</e2>", " ")
    sen = sen.replace("<e1>", " ")
    sen = sen.replace("</e1>", " ")
    
    ce_data.at[index,'sentences']= sen
    
```

```python
ce_data['cause'] = cause
ce_data['effect'] = effect
ce_data['relation'] = relation
```
